```python
import numpy as np
import matplotlib.pyplot as plt

def print_func(func):
  x=np.arange(-5,5,0.1)
  y=func(x)
  plt.plot(x,y)
  plt.show()
```

## 활성화 함수

활성화 함수는 신경망 함수에 비선형성을 주는 역할을 한다.
활성화 함수가 없다면 층을 깊게 쌓는 의미가 없어진다.

### 1. 계단 함수

```python
 def step(x):
   return np.array(x>0, dtype=np.int)
  
print_func(step)
```

### 2. 시그모이드 함수

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

print_func(sigmoid)
```

### 3. tanh 함수

```python
def tanh(x):
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

print_func(tanh)
```

### 4. ReLU

```python
def ReLU(x):
  return np.maximum(0,x)

print_func(ReLU)
```

## 출력층

### 항등 함수

- **회귀**
- 입력 데이터에서 연속적인 수치를 예측
- 출력에 연속성이 있다.

### 소프트맥스

- **분류**
- 입력 데이터에서 어느 클래스에 속하는지 분류
- class가 정해져 있다.
- 각 출력값을 "확률"로 생각할 수 있다.
- 오버플로 주의!


```python
def softmax(x):
  max=np.max(x)
  x=np.exp(x-max) ##오버플로 방지
  sum=np.sum(x)
  return x/sum

print(softmax(np.array([0.3,2.9,4.0])))
```
